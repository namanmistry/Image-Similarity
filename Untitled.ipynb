```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
```

```python
incept_model = ResNet50(include_top=True)
last = incept_model.layers[-2].output
model = Model(inputs= incept_model.input, outputs=last)
```

```python
import cv2
import numpy as np
img = cv2.imread("./data/car_0009.jpg")
img = cv2.resize(img,(224,224))
img_array = np.array(img).flatten()
pred = model.predict(img_array.reshape(1,224,224,3))
pred.shape
```

```python
from scipy import spatial
img = cv2.imread("./data/car_0001.jpg")
img = cv2.resize(img,(224,224))
img_array = np.array(img).flatten()
pred1 = model.predict(img_array.reshape(1,224,224,3))

result = 1 - spatial.distance.cosine(pred, pred1)
result
```

```python
import os
top_10_list = []
top_10_files = []
for file in os.listdir("./data/"):
    try:
        img = cv2.imread(f"./data/{file}")
        img = cv2.resize(img,(224,224))
        img_array = np.array(img).flatten()
        pred1 = model.predict(img_array.reshape(1,224,224,3))
        result = 1 - spatial.distance.cosine(pred, pred1)
        if len(top_10_list) <10:
            top_10_list.append(result)
            top_10_files.append(file)
        print(top_10_list)
        for i in range(len(top_10_list)):
            if result > top_10_list[i]:
                top_10_list[i] = result
                top_10_files[i] = file
                break
    except:
        pass
top_10_list
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
for i in range(10):
    img = mpimg.imread(f"./data/{top_10_files[i]}")
    plt.imshow(img)
    plt.show()
```
